We have seasonal quarterly financial data from Johnson and Johnson 1960-1981 in https://s3-us-west-2.amazonaws.com/dsci/6002/data/jj.txt

1) Plot the time series data.  

```python
import pandas as pd

df = pd.read_csv("https://s3-us-west-2.amazonaws.com/dsci/6002/data/jj.txt")
df.head()
```

```python
%pylab inline
df.plot()
```

2) What do you see in the plot? (Trend, seasonality...)  

The trend is generally increasing, and there is evidence of seasonality becuase the line ocillates at regular intervals.

The period is about 4, and the seasonality has more variance toward the end.

3) BONUS: Modify the data to make it approximately stationary.  

```python
df_diff = df.diff().dropna()
df.plot()
```

4) BONUS: Plot the ACF and PACF of the modified data.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_diff, lags=range(1,80))
```
